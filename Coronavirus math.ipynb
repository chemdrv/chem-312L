## Exploring doubling time

```python
def virus_cases(time, initial, doubling):
    """Returns the number of virus cases over time axis given an initial count and a doubling time"""
    return initial * np.exp(np.log(2)/doubling * time)
```

```python
days = np.arange(0, 90, 1)
```

```python
plt.plot(days, virus_cases(days, 6, 7))
plt.plot(days, virus_cases(days, 6, 6))
plt.plot(days, virus_cases(days, 6, 5))
#plt.plot(days, virus_cases(days, 20, 6))
#plt.plot(days, virus_cases(days, 100, 4))
plt.ylim(0,6000000)
```
